```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from stash import simulate_lightcurve
from astropy.io import fits
```

```python
hdu = fits.open('hmi_twospots/hmi.ic_45s.20130513_185530_TAI.2.continuum.fits', memmap=False)
hdu[1].verify('silentfix')
image = hdu[1].data
```

```python
import astropy.units as u
from astropy.constants import R_sun, R_jup, R_earth
period = 1 * u.day
a = 0.02 * u.AU
b = -0.8

plt.style.use('dark_background')

lc = simulate_lightcurve(image.T, period, a, b, R_planet_physical=0.1 * R_jup)

fig, ax = plt.subplots(figsize=(5, 4))
ax.plot(lc.times, lc.fluxes[::-1], lw=2, color='w')
ax.set_ylabel('Flux', fontsize=20)
ax.set_yticks([])
ax.set_xlabel('Time [d]', fontsize=20)

ax.set_facecolor('k')

for s in ['right', 'top']:
    ax.spines[s].set_visible(False)
    
fig.savefig('misaligned.pdf', bbox_inches='tight')
```
